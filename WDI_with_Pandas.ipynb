```python
# Take a specific df corresponding to the year
df_specific_year = df[df['year'] == tahun]
    
# Create a set of all the indicator names in that specific year
indicator_name_specific_set = set(df_specific_year['indicator_name'].unique())
    
# Subtract all the indicator names set into the specific set ... to get the missing indicators
the_subtract_list = list(indicator_name_set - indicator_name_specific_set)
    
# Iterate all missing indicators .... in order to add them into the df
for indikator in the_subtract_list:
    mod_df = mod_df.append({'year' : tahun, 'indicator_name':indikator, 'indicator_value': 0.0}, ignore_index=True)
```

Let us iterate each `year` and add **missing indicator(s)** to it.

```python
counter = 0
```

```python
for tahun in year_arr:
    # Take a specific df corresponding to the year
    df_specific_year = df[df['year'] == tahun]
    
    # Create a set of all the indicator names in that specific year
    indicator_name_specific_set = set(df_specific_year['indicator_name'].unique())
    
    # Subtract all the indicator names set into the specific set ... to get the missing indicators
    the_subtract_list = list(indicator_name_set - indicator_name_specific_set)
    
    # Iterate all missing indicators .... in order to add them into the df
    for indikator in the_subtract_list:
        mod_df = mod_df.append({'year' : tahun, 'indicator_name':indikator, 'indicator_value': np.nan}, ignore_index=True)
    
    if (counter % 10 == 0):
        print("Processed: year", tahun)
    counter+=1
```
